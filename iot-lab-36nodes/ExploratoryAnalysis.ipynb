```python
import pandas as pd
import sys
sys.path.append('../')
import trace_analysis
```

Parse the 1st trace file with 1000 packets over a network without any attacker.

```python
hop_nodes = trace_analysis.process_iotlab_aggregated('/home/ichatz/Local/iot-netprofiler/iot-lab-36nodes/traces/', '2019-01JAN-29-1')
```

```python
std_values, outliers = trace_analysis.separate_outliers(hop_nodes)
```

Produce Historgram of delay depending on the hop-distan

```python
import matplotlib.pyplot as plt

color = ['', 'blue', 'green', 'red']

fig, ax = plt.subplots(figsize=(18, 8))
for hop in hop_nodes.keys():
    hop_nodes[hop].plot.kde(ax=ax, color=color[hop], label='Hop ' + str(hop) + ' KDE')
    hop_nodes[hop].plot.hist(density=True, color=color[hop], alpha=0.3, ax=ax, label='Hop ' + str(hop) + ' Hist')

ax.set_ylabel('Probability')
ax.set_xlabel('Rount Trip Time (RTT) in milliseconds (ms)')
ax.grid(axis='y')
ax.set_title('Distribution of the Dataset')
ax.set_xlim([-40, 100])
ax.legend()
plt.show()
```
